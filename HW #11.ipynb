Part 1: This function returns the top or bottom player(s) in a given category, filtering for all specified teams and giving values for all specified years.

```python
def nba_leader_by_year(df, cat, top = True, year = list(dfnba["Year"].unique()), team = list(dfnba["Tm"].unique())):
    
    games40 = df[df["G"] >= 40] #Get only players with 40 or more games in the year
    
    if type(year) != list:
        year = [year] #if only one year or team is entered, put it in a list so it can work in for loops
    if type(team) != list:
        team = [team]
        
    for i in list(df["Tm"].unique()): #Ged rid of teams not in our list of specified teams
        if i not in team:
            games40 = games40[games40["Tm"] != i]
    
    top_stat_list = []
    for i in year:
        top_stat_list.append(games40[games40["Year"] == i][["Player", cat, "Year"]])
        
               
    return top_stat_list
    
nba_leader_by_year(dfnba, "PTS", year = [2009, 2014])
```

Part 2: This function plots player age against avaerages all of the statistics inputted by the user for each age. It also returns a table showing correlation values between age and the statistics.

```python
def age_perform(dfnba, stats):
    try:
        for i in stats:
            averages = dfnba.groupby("Age").mean()
            y_vals = pd.DataFrame(averages[i])
            x = list(y_vals.index)
            y = list(y_vals.values.flatten())
            plt.bar(x, y)
            plt.title(f"Player Age vs Average {i} (2008-2017)")
            plt.xlabel("Age")
            plt.ylabel(f"{i}")
            plt.show()
        stats.insert(0, "Age")
        return dfnba[stats].corr()
    except:
        print("No plot available! Please enter a valid numerical statistic from dfnba.")
age_perform(dfnba, ["G", "PTS", "AST"])
```

Part 2 Findings: According to the plots, the strongest relationship between age and our given stats is the relationship between age and total points, as points begin to gradually decline starting at age 28.  However, the correlation matrix shows the strongest overall correlation (involving age) being between age and assists, though this is still very weak.

Part 3.1: These plots compare stats between eastern and western teams.

```python
def east_west_teams(dfnba, stats):
    dfnba["DIV"] = ["West" if x in ['DAL', 'DEN', 'GSW', 'HOU', 'LAC', 'LAL', 'MEM', 'MIN', 'NOP', 'OKC', 'PHX', 'POR', 'SAC', 'SAS', 'UTA'] else "East" for x in dfnba["Tm"]]
    west = dfnba[dfnba["DIV"] == "West"].dropna()
    east = dfnba[dfnba["DIV"] == "East"].dropna()
    for i in stats:
        plt.subplot(1,2,1)
        plt.hist(x = list(west[i].values.flatten()))
        plt.title(f"West Division {i}s")
        plt.xlabel(f"{i}")
        plt.ylabel("Count")
        plt.subplot(1,2,2)
        plt.hist(x = list(east[i].values.flatten()))
        plt.title(f"East Division {i}s")
        plt.xlabel(f"{i}")
        plt.show()
east_west_teams(dfnba, ["G", "AST"])
```

Part 3.2: These plots compare the top 10 western and eastern players for given stats.

```python
def east_west_players(dfnba, stats):
    dfnba["DIV"] = ["West" if x in ['DAL', 'DEN', 'GSW', 'HOU', 'LAC', 'LAL', 'MEM', 'MIN', 'NOP', 'OKC', 'PHX', 'POR', 'SAC', 'SAS', 'UTA'] else "East" for x in dfnba["Tm"]]
    west = dfnba[dfnba["DIV"] == "West"].dropna()
    east = dfnba[dfnba["DIV"] == "East"].dropna()
    for i in stats:
        top_west = west.groupby("Player").mean().sort_values(i, ascending = False)[i].head(10)
        top_east = east.groupby("Player").mean().sort_values(i, ascending = False)[i].head(10)
        x_vals1 = list(top_west.index)
        y_vals1 = list(top_west.values)
        x_vals2 = list(top_east.index)
        y_vals2 = list(top_east.values)
        plt.subplot(1,2,1)
        plt.bar(x_vals1, y_vals1)
        plt.title(f"Top West Players for {i}")
        plt.xlabel("Player")
        plt.ylabel(f"Mean {i}")
        plt.xticks(rotation = 90)
        plt.subplot(1,2,2)
        plt.bar(x_vals2, y_vals2)
        plt.title(f"Top East Players for {i}")
        plt.xlabel("Player")
        plt.xticks(rotation = 90)
        plt.show()
        
        
        
east_west_players(dfnba, ["PTS", "AST"])
```

Part 3 Findings: The top players in the eastern division are stronger overall for assists and total points scored than the top players in the western division.  The top two players in average assists per year in the eastern division have far more assists than the top two in the western division.
