### Plotting $x\sin\left(\frac 1x\right)$
Graph the functions $x\sin\left(\frac 1x\right)$, $|x|$, and $-|x|$ on $[-0.2, 0.2]$ using 2000 points. Use the `np.sin` and `np.abs` functions to produce the $y$ values. No looping is necessary.

<img src="https://spot.colorado.edu/~sschang/python/xsinx-graph.png" />

````python
x_vals = np.linspace(-0.2, 0.2, 2000)
y_vals1 = [i * np.sin(1 / i) for i in x_vals]

y_vals2 = [np.abs(i) for i in x_vals]
y_vals3 = [np.abs(i) * -1 for i in x_vals]

plt.plot(x_vals, y_vals1)
plt.plot(x_vals, y_vals2)
plt.plot(x_vals, y_vals3)
plt.plot
plt.show()
````

### A Histogram
Simulate the rolling of two dice 1000 times and calculate the sum for each roll. Plot the sums as a histogram.

\end{cases}$$
