* **Median:** Median earnings of full-time, year-round workers
* **P25th:** 25th percentile of earnings
* **P75th:** 75th percentile of earnings

```python
college["Total"] = college["Men"] + college["Women"]
college.columns
```

```python
college.sort_values(by = "Total").head(10)["Major"]

#10 Least popular majors
```

```python
college.sort_values(by = "Total", ascending = False).head(10)["Major"]

#10 Most popular majors
```

```python
college[(college["Total"] > 30000) & (college["Total"] < 31000)]

#2 majors
```

```python
x_vals = college.sort_values(by = "Total").head(10)["Major"]
y_vals = college.sort_values(by = "Total").head(10)["Total"]

plt.bar(x_vals, y_vals)
plt.xticks(rotation = 90)
plt.show()
```
