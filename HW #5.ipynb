Problem 1: Plotting the function  ùë¶=ùë•‚àí3sinùë•  and its derivatives  ùë¶‚Ä≤  and  ùë¶‚Ä≥  on  [‚àí4,4]  as line graphs.

```python
import math as m
import matplotlib.pyplot as plt
import random
```

```python
def partition_pts(a,b,n): #helper function to get partition points
    part_diff = abs((b-a) / n)
    part_list = []
    ([part_list.append(a + (part_diff * i)) for i in range(n+1)])
    return part_list

sins_list = partition_pts(-4, 4, 100)
sins_list2 = partition_pts(-4, 4, 50)


def sin_vals1(n): 
    new_list = []
    ([new_list.append(i - (3 * m.sin(i))) for i in sins_list])
    return [round(i,6) for i in new_list]

def sin_vals2(n): 
    new_list = []
    ([new_list.append(1 - (3 * m.cos(i))) for i in sins_list])
    return [round(i,6) for i in new_list]

def sin_vals3(n): 
    new_list = []
    ([new_list.append(3 * m.sin(i)) for i in sins_list])
    return [round(i,6) for i in new_list]

plt.plot(sins_list, sin_vals1(100))
plt.plot(sins_list, sin_vals2(100))
plt.plot(sins_list, sin_vals3(100))
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("y = x - 3sin(x) and its first three derivatives")
plt.show()
```

Problem 3: This function creates a frequency list of random numbers 1-10 in a list of size n, and then plots the distribution.

```python
def random_freq(n):
    num_list = [random.randint(0,9) for i in range(n)]
    freq_list = []
    for i in range(10):
        counter = 0
        for j in range(len(num_list)):
            if num_list[j] == i:
                counter += 1
        freq_list.append(counter)
    return freq_list
nums = list(range(10))

plt.bar(nums, random_freq(1000))
plt.xlabel("Digit")
plt.ylabel("Frequency")
plt.title("Random Frequency Distribution of Digits 0-9")
plt.show()

#Based on the example plot in the HW document, I am assuming I should graph frequencies of the numbers 0 to 9 instead
# of 1 to 10
```

Problem 4: This function returns the value of the normal distribution function for a given value of  ùë•  and parameter sigma, and then plots the function for ùúé=0.5,1.0,1.5  on  [‚àí4,4]. 

```python
def normal(x, sigma):
    ans1 = m.e ** (-(x ** 2) / (2 * (sigma ** 2)))
    ans2 = 1 / (sigma * m.sqrt(2 * m.pi))
    return ans1 * ans2

def partition_pts(a,b,n):
    part_diff = abs((b-a) / n)
    part_list = []
    ([part_list.append(a + (part_diff * i)) for i in range(n+1)])
    return part_list

x_vals = partition_pts(-4,4,100)

y_vals1 = [normal(i, 0.5) for i in x_vals]
y_vals2 = [normal(i, 1) for i in x_vals]
y_vals3 = [normal(i, 1.5) for i in x_vals]


plt.plot(x_vals, y_vals1)
plt.plot(x_vals, y_vals2)
plt.plot(x_vals, y_vals3)
plt.show()
```

Now with 30 values of sigma...

```python
for j in partition_pts(1,4,30):
    plt.plot(x_vals, [normal(i, j) for i in x_vals])
plt.show()
```

Challenge Problem 1: This plot shows the functions  ùë¶=ùë•ùëõsin(ùë•)  for  ùëõ=1,2,3,4 , on the interval  [‚àí10,10] . Each function is scaled to a maximum value of 1.

```python
def chal_func(x_vals, n):
    y_vals = [((x ** n) * m.sin(x)) for x in x_vals]
    max_val = max(y_vals)
    return [y / max_val for y in y_vals]
x_vals = partition_pts(-10,10,100)

for j in range(1,5):
    plt.plot(x_vals, chal_func(x_vals, j))

plt.legend(['$x^1sin(x)$', '$x^2sin(x)$', '$x^3sin(x)$', '$x^4sin(x)$'])
plt.show()
```

Challenge Problem 2: Generates 100 points located at random  ùë•  and  ùë¶  coordinates. Display the points in 4 different subplots, each using a different colormap. The points should be transparent.
